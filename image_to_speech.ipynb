```python
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt
from gtts import gTTS
from IPython.display import Audio
```

```python
ip = "/content/ai hand writing.jpg"
```

```python
plt.imshow(Image.open(ip))
plt.axis("off")
plt.show()
```

```python
img = Image.open(ip)
text = pytesseract.image_to_string(img)
print("Extracted text from the image:\n", text)
```

```python
tts = gTTS(text=text, lang='en')
tts.save("output.mp3")
```
